```python
np_trails = mydata

def recommend_trails(dataset, park_name = None, min_elevation_gain = None, max_elevation_gain = None, min_rating = None, trail_type = None):
  
    if min_elevation_gain is None:
        min_elev = 0
    else:
        min_elev = min_elevation_gain
    if max_elevation_gain is None:
        max_elev = 0
    else:
        max_elev = max_elevation_gain
    if min_rating is None:
        rating = 0
    else:
        rating = min_rating
    if trail_type is None:
        type_t = ""
    else:
        type_t = trail_type
    
    if park_name is None:
        dataset = dataset
    else:
        dataset = dataset[dataset['area_name'] == park_name]
                       
    if type_t == "":
        dataset = dataset[(dataset['elevation_gain'] >= min_elev) 
                          & (dataset['elevation_gain'] >= max_elev) 
                          & (dataset['avg_rating'] >= min_rating)]
    else:
        dataset = dataset[(dataset['elevation_gain'] >= min_elev) 
                          & (dataset['elevation_gain'] >= max_elev) 
                          & (dataset['avg_rating'] >= min_rating)
                          & (dataset['route_type'] == type_t)]
  
    return dataset
```

```python
output = recommend_trails(np_trails, "Haleakala National Park", min_elevation_gain = 1000, min_rating = 4)

print(output[['name', 'elevation_gain', 'avg_rating']])
```

```python
def scrape_month_temp(month = None):
  
    my_url = "https://www.extremeweatherwatch.com/us-state-averages"
    
    if month is None:
        response = requests.get(my_url)
    else:
        response = requests.get(my_url + "/month-" + month)
    soup = BeautifulSoup(response.content, "html.parser")
  
    rows = []

    # iterate over all rows in the faculty table
    for city in soup.find_all("tr")[1:]:
        
        cells = city.find_all("td")
        
        
    
        name_tag = cells[0].find("a") or cells[0]
        name = name_tag.text
        
        max_tag = cells[1].find("td") or cells[1]
        max_t = max_tag.text
        
        min_tag = cells[2].find("td") or cells[2]
        min_t = min_tag.text
        
        rows.append({
        "State": name,
        "max": max_t,
        "min": min_t,
        })
        
    dataframe = pd.DataFrame(rows)
  
  
    return dataframe
```

```python
df = scrape_month_temp(month = "january")
```
